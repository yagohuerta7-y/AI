#### **Importamos las librerías necesarias**

```python
import ollama
# Es una biblioteca de lenguaje diseñada para interactuar con modelos preentrenados. Puede ser utilizada para conversaciones, generación de texto y mucho más.

import re
# Esta biblioteca proporciona herramientas para la manipulación y búsqueda de patrones en cadenas de texto.

import json
# Es una biblioteca que proporciona funciones para codificar y decodificar objetos JSON, que es un formato común para intercambiar datos entre sistemas.

import PyPDF2
# Esta biblioteca permite leer y escribir documentos PDF. Ofrece diversas funciones como extraer texto, cambiar las hojas de papel, etc.

import pymupdf
# Es una versión del lector PDF MuPDF, que ofrece características avanzadas y eficientes para la manipulación de archivos PDF. Es especialmente útil para trabajar con grandes archivos o cuando se requiere un rendimiento óptimo.
```

#### **Creamos las funciones de lectura del PDF**

```python
# Abre el archivo PDF en modo binario para lectura.
pdf_file = open('carta de servicio.pdf', 'rb')

# Crea un objeto PdfReader a partir del archivo abierto.
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Itera sobre cada página del documento.
for page in pdf_reader.pages:
    # Extrae el texto de la página actual utilizando extract_text().
    text = page.extract_text()
    
    # Imprime el texto extraído. Esto se realizará para cada página del PDF.
    print(text)

# Cierra el archivo después de procesar todas las páginas.
pdf_file.close()
```

#### **Conversación con el LLM**

```python
prompt = "Eres un experto en la lectura de documentos, y de generación de resúmenes. Y das la información en formato json"

# Campos que quiero que conteste
fields = ["Título", "Autor", "Tema", "Fecha", "Resumen"]

# phi3.5:latest
# deepseek-r1:latest


# Respuesta
response = ollama.chat(model="phi3.5:latest",
                       messages=[{"role": "system", "content": prompt},
                                 {"role": "user", "content": f"Extrae los siguientes campos {fields} del siguiente documento: {text}"}],

                        options={

                            "temperature":0.2

                        }
                       )

analysis = response["message"]["content"]                
```
